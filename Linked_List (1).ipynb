```python
# ********  MAKING A NODE OF LINKED LIST  ********
class Node:
  def __init__(self,value):
    self.data = value
    self.next = None
```

```python
class LinkedList:
  def __init__(self):
    pass
#  ******  MAKING AN EMPTY LL  ******
    #Empty linked list
    self.head = None
    self.n = 0              # how many nodes in the linked list


    # length of the LL
  def __len__(self):
    return self.n     # number of nodes in the LL



  # traverse/print the elements of LL
  def __str__(self):
    curr = self.head
    result = ''
    while curr != None:
      result = result + str(curr.data) + '->'
      curr = curr.next

    return result[:-2]


# ******  INSERTION IN LL  ********

    # insert from head in LL
  def insert_head(self,value):
    new_node = Node(value)       #making new node by node class
    new_node.next = self.head    #create connection
    self.head = new_node         #reassign head
    self.n += 1


  # insert from tail(append)
  def append(self,value):
    new_node = Node(value)

    if self.head == None:
      self.head = new_node
      self.n = self.n + 1
      return
    curr = self.head

    while curr.next != None:
      curr = curr.next

      # you are at the next node
    curr.next = new_node
    self.n += self.n


  # insert in the middle of LL
  def insert_after(self,after,value):

    new_node = Node(value)
    curr = self.head

    while curr != None:
      if curr.data == after:
        break
      curr = curr.next

    if curr != None:
      new_node.next = curr.next
      curr.next = new_node
      self.n = self.n + 1
    else:
      return  'Item not found'



# ******  DELETION IN LL  ********
  # clear the LL
  def clear(self):
    self.head = None
    self.n = 0
    return 'List is empty'


  #delete item from first/head
  def delete_head(self):

    if self.head == None:
      return 'List is empty'

    self.head = self.head.next

  # delete item from last/tail
  def pop(self):
    if self.head == None:
      return 'List is empty'

    curr = self.head
    #if LL contains only 1 item
    if curr.next == None:
      self.delete_head()

    while curr.next.next != None:
      curr = curr.next

    curr.next = None
    self.n = self.n - 1


  #deleting a specific item
  def remove(self,value):

    if self.head == None:
      return 'List is empty'

    if self.head.data == value:
      return self.delete_head()

    curr = self.head

    while curr.next != None:
      if curr.next.data == value:
        break
      curr = curr.next

    if curr.next == None:
      return 'Item not found'
    else:
      curr.next = curr.next.next



# *****  SEARCHING IN LL  *****
  def search(self,item):

    curr = self.head
    pos = 0

    while curr != None:
      if curr.data == item:
        return pos
      curr = curr.next
      pos = pos + 1

    return 'Not found'
  # for using indexing feature
  def __getitem__(self,index):
    curr = self.head
    pos = 0

    while curr != None:
      if pos == index:
        return curr.data
      curr = curr.next
      pos = pos + 1

    return 'IndexError'

    # replace max value with the given value in LL
  def replace_max(self,value):
    temp = self.head
    max = temp

    while temp != None:
      if temp.data > max.data:
        max = temp
      temp = temp.next

    max.data = value


  # sum of odd postion's value from the LL
  def sum_odd_nodes(self):
    temp = self.head
    counter = 0
    result = 0
    while temp != None:
      if counter%2 != 0:
        result = result + temp.data
      counter += 1
      temp = temp.next
    print(result)


  # to reverse a LL / inplace reversal
  def reverse(self):
    prev_node = None
    curr_node = self.head

    while curr_node != None:
      next_node = curr_node.next
      curr_node.next = prev_node
      prev_node = curr_node
      curr_node = next_node

    self.head = prev_node

  # replace * or / by a single space
  # if two consecutive occurence ** or // then replace by single space and convert next char to uppercase
  def change_sent(self):
    temp = self.head
    while temp != None:
      if temp.data == '*' or temp.data == '/':
        temp.data = ' '
      temp = temp.next
```

```python
L = LinkedList()
L.insert_head(1)
L.insert_head(4)
L.insert_head(5)
L.append(9)
L.insert_after(5,100)
```
